```python
#Importing pymongo

import pymongo
```

```python
#client Mongo

client = pymongo.MongoClient("localhost", 27017)
```

```python
#Database names

client.list_database_names()
```

```python
#Setting db as foodsales

db = client ["foodsales"]
```

```python
#Listing out collection names

db.list_collection_names()
```

```python
#Counting the no of documents

db.abc.count_documents({ })
```

```python
#Finding one document using findone()

db.abc.find_one()
```

```python
db.abc.count_documents({})
```

```python
#Displaying the documents

for i in db.abc.find():
    print(i)
```

```python
#Finding the Type

type(db.abc.find())
```

```python
#Documents with not displaying id

for i in db.abc.find({},{"_id":0}):
    print(i)
```

```python
#Displaying the documents specified to the east region

for i in db.abc.find({'Region':'East'},{"_id":0,"Category":1,"Product":1,"City":1,"TotalPrice":1,"Region":1}):
    print(i)
```

```python
x = db.abc.find({},{"_id":0,"Category":1,"Product":1,"City":1,"TotalPrice":1,"Region":1}).sort("TotalPrice", -1).limit(5)

for i in x:
    print(i)
```

```python
x = db.abc.find({"TotalPrice": { "$gt": 500 }},{"_id":0,"Category":1,"Product":1,"City":1,"TotalPrice":1,"Region":1})

for i in x:
    print(i)
```

```python
#Displaying the low 5 products

x = db.abc.find({},{"_id":0,"Category":1,"Product":1,"City":1,"TotalPrice":1,"Region":1}).sort("TotalPrice").limit(5)

for i in x:
    print(i)              
```

```python
#Displaying the Top 5 products

x = db.abc.find({},{"_id":0,"Category":1,"Product":1,"City":1,"TotalPrice":1,"Region":1}).sort("TotalPrice",-1).limit(5)

for i in x:
    print(i)     
```

```python
#Top 5 products in East Region

x = db.abc.find({"Region":'East'},{"_id":0,"Category":1,"Product":1,"City":1,"TotalPrice":1,"Region":1}).sort("TotalPrice",-1).limit(5)

for i in x:
    print(i)  
```
